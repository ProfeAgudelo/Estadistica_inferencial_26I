```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import t
from scipy.stats import chi2
```

#Ejemplo: Interalo de confianza para la media

##
**Citas rápidas**  En un estudio sobre citas rápidas realizado en la Universidad de Columbia, se pidió a los sujetos femeninos que caliﬁ caran el atractivo de sus citas masculinas, y una muestra de los resultados se detalla a continuación (1 = no atractivo; 10 = extremadamente atractivo). Use un nivel de conﬁanza del 99%. ¿Puede usarse el resultado para estimar el atractivo medio de la población de todos los adultos masculinos?

```python
datos_atractivo = [5, 8, 3, 8, 6, 10, 3, 7, 9, 8, 5, 5, 6, 8, 8, 7, 3, 5, 5, 6, 8, 7, 8, 8, 8, 7]
```

```python
#1. Cálculo de medidas de tendencia central y valor de n
media = np.mean(datos_atractivo)
print (f'La media de los datos es: {media:.4f}')

dev_est = np.std(datos_atractivo, ddof=1) #Desviación estandar muestral
print (f'La desviación estándar de los datos es: {dev_est:.4f}')

n = len(datos_atractivo)
print (f'El número de datos es: {n}')
```

```python
#2. Cálculo del valor de alfa y de ji cuadrado.
c= 0.99 #Valor de la confianza
alfa_medio = (1 - c)/2  #Valor de alfa

valor_t = t.ppf(1 - alfa_medio, n - 1)
print (f'El valor de t es: {valor_t:.4f}')
```

```python
#3. Calcular el Margen de Error y el Intervalo
margen_error = valor_t * (dev_est / np.sqrt(n))
limite_inferior = media - margen_error
limite_superior = media + margen_error

print(f'Margen de error (E): {margen_error:.4f}')
print(f'Intervalo de confianza del 99%: ({limite_inferior:.4f}, {limite_superior:.4f})')
```

#Ejemplo: Interalo de confianza para la desviación estandar.

##
**Citas rápidas** En un estudio sobre citas rápidas realizado en la Universidad de Columbia, se pidió a las mujeres que calificaran el atractivo de sus citas masculinas, y una muestra de los resultados se detalla a continuación (1 = no atractivo; 10 = extremadamente atractivo). Construya una estimación del intervalo de confianza del 95% para la desviación estándar de la población a partir de la cual se obtuvo la muestra.

```python
datos_atractivo = [5, 8, 3, 8, 6, 10, 3, 7, 9, 8, 5, 5, 6, 8, 8, 7, 3, 5, 5, 6, 8, 7, 8, 8, 8, 7]
```

```python
dev_est = np.std(datos_atractivo, ddof=1) #Desviación estandar muestral
print (f'La desviación estándar de los datos es: {dev_est:.4f}')

n = len(datos_atractivo)
print (f'El número de datos es: {n}')
```

```python
#2. Cálculo del valor de alfa y de ji cuadrado.
c= 0.95 #Valor de la confianza
alfa_medio = (1 - c)/2  #Valor de alfa

valor_ji_der = chi2.ppf(alfa_medio,n-1)
valor_ji_izq = chi2.ppf(1 - alfa_medio, n-1)
print (f'El valor de ji cuadrado en la derecha es: {valor_ji_der:.4f}')
print (f'El valor de ji cuadrado en la izquierda es: {valor_ji_izq:.4f}')
```

```python
limite_inferior = np.sqrt((n - 1) * dev_est**2 / valor_ji_izq)
limite_superior = np.sqrt((n - 1) * dev_est**2 / valor_ji_der)
print(f'Intervalo de confianza del 95%: ({limite_inferior:.4f}, {limite_superior:.4f})')
```
