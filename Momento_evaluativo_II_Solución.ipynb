```python
#Librerias a emplear
from scipy.stats import binom
from scipy.stats import poisson
from scipy.stats import norm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

En un examen estandarizado de competencia lingüística internacional (tipo IELTS), los puntajes han sido transformados a una escala $Z$ (Distribución normal estandarizada).

Genere un script que calcule y proporcione la visualización para:


*  ¿Qué porcentaje de estudiantes obtiene una calificación $Z$ superior a $1.64$? (Debe mostrar la gráfica de la distribución con el área sombreada correspondiente).
*  ¿Cuál es el valor $Z$ que delimita al 10\% de los estudiantes con peores resultados? (Debe mostrar la gráfica que representa esta situación).

```python
z = 1.64
prob_normal = norm.cdf(z)
prob_complementaria = 1 - prob_normal
print(f"La probabilidad de que estudiantes obtegan un puntaje mayor a: {z} es: {prob_complementaria:.4f}")


# Crear datos para el eje X
x = np.linspace(-4, 4, 100)
y = norm.pdf(x, 0, 1) # Función de densidad

plt.plot(x, y)
plt.title("Distribución Normal Estándar")
plt.xlabel("Puntaje Z")
plt.ylabel("Densidad")
plt.fill_between(x, y, where=(x >= z), color='skyblue', alpha=0.5)
plt.show()
```

```python
area_objetivo = 0.10 #Probabilidad del problema
z_limite = norm.ppf(area_objetivo)
print(f"El valor Z para el {area_objetivo} de estudiantes con peores resultados es de: {z_limite:.4f}")


# Crear datos para el eje X
x = np.linspace(-4, 4, 100)
y = norm.pdf(x, 0, 1) # Función de densidad

plt.plot(x, y)
plt.title("Distribución Normal Estándar")
plt.xlabel("Puntaje Z")
plt.ylabel("Densidad")
plt.fill_between(x, y, where=(x < z_limite), color='skyblue', alpha=0.5)
plt.show()
```
