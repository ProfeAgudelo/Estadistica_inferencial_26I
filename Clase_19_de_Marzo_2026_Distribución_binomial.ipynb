# Distribución Binomial

Para que un experimento pueda ser modelado bajo esta distribución, debe cumplir con los denominados **4 Pilares del Modelo**:

1. **Número fijo de ensayos ($n$):** El experimento se repite una cantidad determinada de veces.
2. **Resultados dicotómicos:** Solo existen dos resultados posibles: **"Éxito"** (el evento de interés) o **"Fracaso"**.
3. **Probabilidad constante ($p$):** La probabilidad de éxito no cambia de un ensayo a otro.
4. **Independencia:** El resultado de un ensayo no afecta el resultado de los demás.

### La Fórmula de Probabilidad Binomial

La probabilidad de obtener exactamente $x$ éxitos en $n$ ensayos independientes está dada por:

$$P(x) = \frac{n!}{(n - x)! x!} \cdot p^x \cdot q^{n-x}$$

**Definición de variables:**
* $n$: Número total de ensayos.
* $x$: Número de éxitos deseados.
* $p$: Probabilidad de éxito en cada ensayo.
* $q$: Probabilidad de fracaso ($1 - p$).

La **Función de Masa de Probabilidad** (PMF, por sus siglas en inglés) es una herramienta estadística que asigna la probabilidad exacta de que una variable aleatoria discreta tome un valor específico

```python
#Probabilidad de masa (puntual)
n_ensayos = 15
prob_exito = 0.08
x = 3

prob_binomial = binom.pmf(x,n_ensayos,prob_exito)
print(f'El resultado de la probabilidad binomial es de :{prob_binomial}')
```

La **Función de Distribución Acumulada** (Cumulative Distribution Function), denotada como $F(x)$, calcula la probabilidad de que una variable aleatoria $X$ tome un valor menor o igual a $x$.

1. Definición Matemática
$$F(x) = P(X \leq x)$$

2. Propiedades Clave
* **Rango:** $0 \leq F(x) \leq 1$.
* **Límites:** $\lim_{x \to -\infty} F(x) = 0$ y $\lim_{x \to \infty} F(x) = 1$.
* **Cálculo de intervalos:** $P(a < X \leq b) = F(b) - F(a)$.

```python
#Probabilidad acumulada
n_ensayos = 15
prob_exito = 0.08
x = 0

prob_binomial = binom.cdf(x,n_ensayos,prob_exito)
print(f'El resultado de la probabilidad binomial es de :{prob_binomial}')

#Cómo es al menos uno, calculo la probabilidad complementaria
prob_complementaria = 1 - prob_binomial
print(f'La probabilidad de obtener al menos {x+1} es de: {prob_complementaria}')
```

```python
#Tambien puedo cálcular el al menos x valor, a partir del uso de intervalos
n_ensayos = 15
prob_exito = 0.08
x_1 = 0 #Intervalo 1
x_2 = 15 #Intevalo 2

prob_binomial_1 = binom.cdf(x_1,n_ensayos,prob_exito) #Probabilidad del intervalo 1
prob_binomial_2 = binom.cdf(x_2,n_ensayos,prob_exito) # Probabilidad del intevalo 2
prob_binomial = prob_binomial_2 - prob_binomial_1 # Cálcuilo del intervalo

print(f'El resultado de la probabilidad binomial es de :{prob_binomial}')
```
