```python
#Librerias a emplear
from scipy.stats import binom
from scipy.stats import poisson
from scipy.stats import norm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

##Punto #1
## 1. Ingeniería Industrial - Control de Calidad

Un lote de microcontroladores para automatización tiene una probabilidad histórica del **4%** de presentar fallos de fábrica. Un ingeniero de calidad selecciona una muestra aleatoria de **25 unidades**.

Calcule:

a) La probabilidad de encontrar exactamente **2 defectuosos**.  

b) La probabilidad de encontrar **3 o más defectuosos**.

```python
n = 25 #Muestra de microcontroladores
p = 0.04 #Probabilidad de fallos
x = 2 #Cantidad de microcontroladores defecutosos

prob_binomial = binom.pmf(x,n,p)

print(f'La probabilidad de encontrar {x} sensores defectuosos es de: {prob_binomial:4f}')
```

```python
x=2
prob_binomial = binom.cdf(x,n,p)
#Cómo es al menos dos, calculo la probabilidad complementaria
prob_complementaria = 1 - prob_binomial
print(f'La probabilidad de obtener {x+1} o más sensores defectuosos es de: {prob_complementaria:4f}')
```

##Punto #2
## 2. Ingeniería Multimedia - Rendimiento de Servidores

Un servidor de streaming de video renderiza texturas en tiempo real. La probabilidad de que un *frame* experimente caída de resolución por latencia es del **15%**. Se analiza una ráfaga de **40 frames**.

¿Cuál es la probabilidad de que, como máximo, **5 frames** presenten este problema?

```python
n = 40 #Número de frames
p = 0.15 #Probabilidad éxito
x = 5 #Casos a evaluar

prob_binomial = binom.cdf(x,n,p)

print(f'La probabilidad de encontrar máximo {x} frames defectuosos es de: {prob_binomial:4f}')
```

Un modelo de Procesamiento de Lenguaje Natural (NLP) tiene una tasa de acierto del **88%** al identificar correctamente modismos regionales en traducciones de inglés a español.

Determine la probabilidad de que el modelo traduzca correctamente **exactamente 12 oraciones**.

##Punto #4
## 4. Ingeniería Industrial - Mantenimiento Predictivo

En una línea de ensamblaje automatizada, los brazos robóticos requieren calibración a una tasa promedio de **2.5 veces por turno de 8 horas**.

Calcule:

a) La probabilidad de que **no se requiera ninguna calibración**.  

b) La probabilidad de que se requieran **exactamente 4 calibraciones**.

```python
lambda_poisson = 2.5 #Valor promedio en el tiempo
k_1 = 0 #Valor de evaluación
k_2 = 4 #Valor de evaluación

prob_poisson = poisson.pmf(k_1,lambda_poisson)
print(f'La probabilidad de {k_1} calibraciones es de: {prob_poisson:4f}')

prob_poisson = poisson.pmf(k_2,lambda_poisson)
print(f'La probabilidad de {k_2} calibraciones es de: {prob_poisson:4f}')
```

##Punto #5
## 5. Ingeniería Multimedia - Tráfico de Red

Durante el lanzamiento de una aplicación interactiva, el servidor principal registra un promedio de **6 peticiones por minuto**.

¿Cuál es la probabilidad de que en un minuto específico se reciban **más de 8 peticiones**?

```python
lambda_poisson = 6 #Peticiones por mínuto
k_1 = 8 #Peticiones en un mínuto.

prob_poisson = poisson.cdf(k_1,lambda_poisson)
prob_complementaria = 1 - prob_poisson
print(f'La probabilidad de recibir más de {k_1} peticiones es de: {prob_complementaria:4f}')
```

#Punto #7
## 7. Ingeniería Industrial - Resistencia de Materiales

La resistencia a la tracción de un lote de cables de acero se ha estandarizado (valores Z).

Calcule la probabilidad de que un valor esté entre:

$$
Z = -1.25 \quad \text{y} \quad Z = 0.85
$$

Además, genere la gráfica del área bajo la curva normal estándar.

```python
z_inf = -1.25
z_sup = 0.85
area_entre = norm.cdf(z_sup) - norm.cdf(z_inf)
print(f"El área entre {z_inf} y {z_sup} (Areas de probabilidad para el lote) es: {area_entre:.4f}")

x = np.linspace(-4, 4, 400)
y = norm.pdf(x, 0, 1)

plt.plot(x, y)
plt.title("Distribución Normal Estándar")
plt.xlabel("Puntaje Z")
plt.ylabel("Densidad")

# Se deben usar paréntesis y el operador & para condiciones compuestas en arreglos
plt.fill_between(x, y, where=(x >= z_inf) & (x <= z_sup), color='skyblue', alpha=0.5)

plt.show()
```

#Punto #8
## 8. Ingeniería Multimedia - Latencia en Realidad Virtual

El tiempo de latencia se transformó a escala Z.

Determine la probabilidad de que:


$$Z > 1.64$$


Represente gráficamente esta región.

```python
z = 1.64
probabilidad = norm.cdf(z)
print(f"La probabilidad acumulada para valores mayores (o latencias mayores a:) Z={z} es: {1-probabilidad:.4f}")

# Crear datos para el eje X
x = np.linspace(-4, 4, 100)
y = norm.pdf(x, 0, 1) # Función de densidad

plt.plot(x, y)
plt.title("Distribución Normal Estándar")
plt.xlabel("Puntaje Z")
plt.ylabel("Densidad")
plt.fill_between(x, y, where=(x >= z), color='skyblue', alpha=0.5)
plt.show()
```

## 9. Lenguas - Competencia Lingüística

Los puntajes de una prueba han sido normalizados (valores Z).

Determine la proporción de estudiantes con:


$Z < -0.5$


Muestre gráficamente esta región.

```python
z = -0.5
probabilidad = norm.cdf(z)
print(f"La probabilidad acumulada para estudiantes, que sea menor a Z={z} es: {probabilidad:.4f}")

# Crear datos para el eje X
x = np.linspace(-4, 4, 100)
y = norm.pdf(x, 0, 1) # Función de densidad

plt.plot(x, y)
plt.title("Distribución Normal Estándar")
plt.xlabel("Puntaje Z")
plt.ylabel("Densidad")
plt.fill_between(x, y, where=(x < z), color='skyblue', alpha=0.5)
plt.show()
```
